```python
import os
import numpy as np

data_path="train_data_2"

labels=os.listdir(data_path)
categories=np.arange(len(labels))
category_dict=dict(zip(labels,categories))

print("lables=",labels)
print("categories=",categories)
print("category_dict=",category_dict)
```

```python
import cv2

face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

data=[]
target=[]

for label in labels:#each folder
    imgs_path=os.path.join(data_path,label)
    img_names=os.listdir(imgs_path)#converts image names into an array
    
    for img_name in img_names:#each image in a folder
        img_path=os.path.join(imgs_path,img_name)#path of every image
        img=cv2.imread(img_path)#loads image into an array
        #to display loaded images
        #cv2.imshow('LIVE',img)
        #key=cv2.waitkey(1000) time gap between each image
        #if(key==27) escape key to stop slide show
            #break
    #cv2.destroyAllWindows()
        #preprocessing starts
        #convert colour images to grey
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        #using haarcascade classifier to detect faces
        faces=face_classifier.detectMultiScale(gray)
        
        for (x,y,w,h) in faces:
            cropped_face=gray[y:y+h,x:x+w]
            cv2.imshow('LIVE',cropped_face)#showing cropped faces
            key=cv2.waitKey(0)
            if(key==121):#press y if img is accepted
                cropped_face=cv2.resize(cropped_face,(50,50))
                data.append(cropped_face)
                target.append(category_dict[label])
                print('image is accepted')
            else:
                pass
cv2.destroyAllWindows()
```

```python
data=np.array(data)
target=np.array(target)

noImages,height,width=data.shape

data=data.reshape(noImages,height*width)#faltten

np.save('data',data)
np.save('target',target)
```
