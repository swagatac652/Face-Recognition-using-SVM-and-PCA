# DATA PRE-PROCESSING

## 1. Importing Libraries

```python
import os #used to access files
import numpy as np
import cv2 #OpenCV library used for preprocessing image data
```

## 2. Importing images

```python
data_path="train_data_2"

labels=os.listdir(data_path) #loading the labels of the data present in the folder
categories=np.arange(len(labels)) #generating keys for labels
category_dict=dict(zip(labels,categories)) #converting into key: labels dictionary

print("lables=",labels)
print("categories=",categories)
print("category_dict=",category_dict)
```

## 3. Loading and pre-processing images

```python
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml') #loading haarcascade face classfier

data=[]
target=[]

#loading images
for label in labels:#each folder
    imgs_path=os.path.join(data_path,label) #creates a new path to get inside each folder ie data_path\folder_name
    img_names=os.listdir(imgs_path) #converts image names into an array
    #print(img_names)
    
    for img_name in img_names: #each image in a folder
        img_path=os.path.join(imgs_path,img_name) #path of every image
        #print(img_path)
        img=cv2.imread(img_path) #loads each image into an array
        
        #to display loaded images
        #cv2.imshow('LIVE',img)
        #cv2.imshow('GRAY',gray) to view the images in grayscale
        #key=cv2.waitkey(1000) time gap between each image
        #if(key==27): escape key to stop slide show
            #break
        #cv2.destroyAllWindows()
        
        #preprocessing starts
        #convert colour images to grey
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        #using haarcascade classifier to detect faces from the gray images
        faces=face_classifier.detectMultiScale(gray) #array of dimensions x,y,w,h if the picture has multiple faces? detects someother things as faces
        
        for (x,y,w,h) in faces: #loop through all ROI parameters
            
            #lets draw a rectangle on each ROI
            #cv2.rectange(img,(x,y),(w+x,h+y),(0,255,0),2) #image,point1,point2,color of rectangle,thickness of rectange
            #cv2.imshow('LIVE',img)
            #cv2.waitKey(1000)
            
            cropped_face=gray[y:y+h,x:x+w]
            cv2.imshow('LIVE',cropped_face)#showing cropped faces and wrong cropped items
            key=cv2.waitKey(0)#wait till a key is pressed
            if(key==121):#press y if img is accepted
                cropped_face=cv2.resize(cropped_face,(50,50)) #make a common size for all cropped faces
                data.append(cropped_face) #add correct images to the dataset
                target.append(category_dict[label]) #add corresponing name of the image
                print('image is accepted')
            else:
                pass #donot take into the dataset
cv2.destroyAllWindows()
```

## 4. Save the dataset

```python
data=np.array(data) #convert into numpy 2Darray
target=np.array(target)

noImages,height,width=data.shape # returns number of images, height of image, width of image

data=data.reshape(noImages,height*width) #faltten the data into 1D

np.save('data',data)
np.save('target',target)
```
